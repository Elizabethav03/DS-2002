Data:

```python
data = {
    "Term Start": [1917, 1919, 1921, 1923, 1925, 1927, 1929, 1931, 1933, 1935, 1937, 1939, 1941, 1943, 1945, 1947, 1949, 1951, 1953, 1955, 1957, 1959, 1961, 1963, 1965, 1967, 1969, 1971, 1973, 1975, 1977, 1979, 1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999, 2001, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017],
    "Term End": [1919, 1921, 1923, 1925, 1927, 1929, 1931, 1933, 1935, 1937, 1939, 1941, 1943, 1945, 1947, 1949, 1951, 1953, 1955, 1957, 1959, 1961, 1963, 1965, 1967, 1969, 1971, 1973, 1975, 1977, 1979, 1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999, 2001, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017, 2019],
    "Women Representatives": [1, 1, 4, 2, 3, 5, 9, 8, 8, 6, 6, 9, 9, 9, 10, 10, 9, 10, 11, 17, 16, 16, 17, 12, 11, 11, 10, 13, 16, 19, 18, 16, 22, 22, 23, 24, 29, 32, 48, 50, 55, 58, 60, 62, 66, 74, 77, 73, 79, 84, 89],
    "Total Representatives": [435] * 51
}

df = pd.DataFrame(data)
```

Build the bar graph below:

```python
# Create a new figure with a specific size (10 inches wide by 6 inches tall).
plt.figure(figsize=(10, 6))

# Calculate remaining seats to reach 90
df['Remaining to 90'] = 90 - df['Women Representatives']

# Create stacked bar chart - orange bars first (bottom), then gray bars on top
plt.bar(df['Term Start'], df['Women Representatives'], color='#fe5a03', zorder=2, label='Women Representatives')
plt.bar(df['Term Start'], df['Remaining to 90'], bottom=df['Women Representatives'],
        color='#d3d3d3', alpha=0.7, zorder=1, label='Remaining to 90')

# Add a subtitle to describe what the chart shows.
plt.suptitle('Women Representation in U.S. Congress Over Time', x=0.32, weight='bold')

# Use the title as the y-axis label to indicate the unit of measurement.
plt.title('Number of Women Representatives', x=0.035, size=10)

# Rotate the x-axis labels slightly for better readability.
plt.xticks(rotation=45, ha='right')

# Add a grid to the background for easier comparison of bar heights.
plt.grid(axis='y', linestyle='--', color='gray', which='both', zorder=0)

# Customize the y-axis tick marks to appear at intervals of 20 units.
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(20))

# Remove the top and right borders (spines) of the plot for a cleaner appearance.
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Optionally remove the left border (spine) for an even cleaner look.
plt.gca().spines['left'].set_visible(False)

# Remove the tick marks (lines) from the x-axis and y-axis to declutter the chart.
plt.gca().tick_params(axis='x', which='both', length=0)  # x-axis tick lines removed
plt.gca().tick_params(axis='y', which='both', length=0)  # y-axis tick lines removed

# Add a text box at the bottom left of the figure to credit the data source.
plt.figtext(0.1, 0.0, 'Source: U.S. Congress Historical Data', ha='left', fontsize=10)

# Finally, display the plot on the screen.
plt.tight_layout()
plt.show()
```

Scaffolding of the graph helps tell a story by showing how the orange is the amount filled and yet through all these, 90 seats have not even been reached during all this time. The x-axis also makes it easy to follow as it just goes from oldest to most recent time.
